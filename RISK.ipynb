```python
def main():
    num_rondas = 1
    partida = Risk()
    while num_rondas!=0:
        print(f'Empieza la ronda {num_rondas}')
        partida.dado_Jugador()
        partida.dado_CPU()
        partida.Ataque()
        num_rondas = partida.rondas(num_rondas)
                
if __name__ == '__main__':
    main()
```
